Python functions have several useful features that make them flexible and powerful. Two of these features are default values and named arguments.

**Default Values:** When defining a function in Python, you can assign default values to the arguments. If a value for that argument is not provided when the function is called, the default value is used instead.


```python
def greet(name, greeting="Hello"):
    print(greeting, name)

greet("Alice")
greet("Bob", "Hi")
```

**Named Arguments:** When calling a function in Python, you can specify the argument values by name, rather than by position. This can be especially useful when a function has many arguments, and you want to provide values only for certain ones.

```python
def calculate(a, b, c):
    return a + b - c

print(calculate(1, 2, 3))
print(calculate(a=1, b=2, c=3))
print(calculate(b=2, c=3, a=1))
print(calculate(b=2, a=1, c=3))
```

## Multiple choice

In Python, you can use a multiple choice pattern or a switch statement using the `if`, `elif`, and `else` statements. This pattern allows you to execute different blocks of code based on the value of a variable or expression.

```python
def calculator(num1, num2, operator):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        # add some more lines of code
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 == 0:
            print("Error: cannot divide by zero!")
            return None
        else:
            return num1 / num2
    else:
        print("Error: invalid operator!", operator)
        return None
```

```python
print (calculator(1,2,'/'))
print (calculator(1,2,'f'))
```

# Loops

There are two fundamental programming constructs that allow you to control the flow of a program: conditions and loops. We have already see above the conditional statements that are used to execute different blocks of code depending on whether a certain condition is satisfied.

Loops are used to execute a block of code repeatedly
the most typical and useful loop in `python` is `for`.

`for` loops iterate over a sequence of elements, such as a list or a string.

```python
lst = [4, 8, 0, 9, 5, 9]
for elem in lst: #<-- Beware of the colon mark
  #all lines of code inside the for loop
  print(elem) #<-- the block repeated for every element in the sequence
```

```python
for char in "Hello, good morning!":
  print(char)
```

An application example

```python
def double(lst):
  newlst = []
  for a in lst:
    newlst.append(2*a)
  return newlst
```

```python
lst = [4, 8, 0, 9, 5, 9]
print(double(lst))
```

## **Exercise**

Given a dictionary with the structure of `employee_data`, write functions to find solutions to these questions:

* What is the highest salary?
* What department has more people?


```python
def highest_salary(dict):
    max_salary = 0
    for name in dict:
        salary = dict[name][1]
        if salary > max_salary:
            max_salary = salary
    return max_salary

```

```python
highest_salary(employee_data)
```

```python
def larger_dept(dict):
    department_count = {}
    for _, (department, _) in dict.items():
        if department in department_count:
            department_count[department] += 1
        else:
            department_count[department] = 1
    max_department = max(department_count, key=department_count.get)
    return max_department

larger_dept(employee_data)
```
