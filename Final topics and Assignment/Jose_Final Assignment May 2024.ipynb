```python

#Select all trips that begin or end in Central Park.
df1[df1['end station name'].str.contains('Central Park') | df1['start station name'].str.contains('Central Park')]


```

```python
#Draw charts to study the trip duration for different categories of users

a1=sns.barplot(data=df1, x="usertype", y="tripduration", estimator='mean')

a2=sns.barplot(data=df1, x="usertype", y="tripduration", estimator='median')

a1.set(xlabel='user type', ylabel='Trip duration in minutes', title='Mean duration type')


#a2.set(xlabel='user type', ylabel='Trip duration in minutes', title='Median duration type')

```

```python
#Compare the population size in the different categories

df1.gender.value_counts().plot(kind='pie',
                             autopct='%.1f%%',
                             colormap='summer',
                             title='% of bike users by sex',
                             ylabel='gender',
                             figsize=(6,6),
                             labels=['man', 'women', 'no info'])

df1.usertype.value_counts().plot(kind='pie',
                             autopct='%.1f%%',
                             colormap='summer',
                             title='% of bike users by user type',
                             ylabel='user type',
                             figsize=(6,6),
                             labels=['customer', 'suscribe'])
```
